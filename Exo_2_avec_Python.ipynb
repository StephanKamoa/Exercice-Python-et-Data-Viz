```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

 Un diagramme de disperssion de la colonne prix par rapport Ã  la colonne carat 

```python

sns.scatterplot(x="carat", y="price", data=diamonds)
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

### Utilisation du paramÃ¨tre alpha et le paramÃ¨tre edgecolor pour traiter le problÃ¨me de 

```python
sns.scatterplot(x="carat", y="price", data=diamonds, alpha=0.1, edgecolor='blue')
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
plt.figure(figsize=(10, 8))

sns.scatterplot(x="carat", y="price", data=diamonds, alpha=0.5, edgecolor='blue')
plt.xlabel("carat")
plt.ylabel("price")
plt.show()
```

```python
sns.histplot(data=diamonds, x="price")
plt.xlabel('price')
plt.show()
```

```python
sns.countplot(x="cut", data=diamonds)
plt.xlabel("cut")
plt.ylabel("count")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(x="cut", y="price", data=diamonds)
plt.xlabel("cut")
plt.ylabel("price")
plt.show()
```

```python
array = ['Fair', 'Ideal', 'Good', 'Very Good', 'Premium']
plt.figure(figsize=(10, 8))
sns.boxplot(x="cut", y="price", data=diamonds, palette="pastel", order=array)
plt.xlabel("cut")
plt.ylabel("price")
plt.show()
```
